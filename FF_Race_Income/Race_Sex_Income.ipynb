```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
datafile = "../Data/mrgdata_food.csv"
```

```python
df = pd.read_csv(datafile)
```

```python
df_inc_race = (
    pd.DataFrame(
        [
            df["Median Income (dollars)"],
            df["White Population"],
            df["Black or African American Population"],
            df["American Indian and Alaska Native Population"],
            df["Asian Population"],
            df["Native Hawaiian and Other Pacific Islander population"],
            df["Some Other Race Population"],
        ]
    )
    .transpose()
    .replace("-", 0)
    .astype("int64")
    .sort_values(by=["Median Income (dollars)"], ascending=True)
)
```

```python
df_inc_race.head()
```

```python
correlation = st.pearsonr(
    df_inc_race["Median Income (dollars)"], df_inc_race["White Population"]
)
print(f"The correlation between both factors is {round(correlation[0],2)}")


from scipy.stats import linregress

(slope, intercept, rvalue, pvalue, stderr) = linregress(
    df_inc_race["Median Income (dollars)"], df_inc_race["White Population"]
)
regress_values = [
    income * slope + intercept for income in df_inc_race["Median Income (dollars)"]
]
line_eq = "y = " + str(slope) + "x + " + str(round(intercept, 2))
# line_eq
print(line_eq)
plt.scatter(df_inc_race["Median Income (dollars)"], df_inc_race["White Population"])
plt.plot(df_inc_race["Median Income (dollars)"], regress_values, "r-")
plt.title("Median Income vs. White Population")
plt.xlabel("Median Income ($)")
plt.ylabel("White Population")
plt.savefig("med_inc_v_white.jpg")
plt.show()
```

```python
correlation_male = st.pearsonr(
    df_inc_gender["Median Income (dollars)"], df_inc_gender["Total Male"],
)
print(
    f"The correlation between male population and median income is {round(correlation_male[0],2)}"
)


from scipy.stats import linregress

(slope, intercept, rvalue, pvalue, stderr) = linregress(
    df_inc_gender["Median Income (dollars)"], df_inc_gender["Total Male"]
)
regress_values_m = [
    income * slope + intercept for income in df_inc_gender["Median Income (dollars)"]
]
line_eq = "y = " + str(slope) + "x + " + str(round(intercept, 2))
# line_eq
print(line_eq)

print()

correlation_female = st.pearsonr(
    df_inc_gender["Median Income (dollars)"], df_inc_gender["Total Female"],
)
print(
    f"The correlation between female population and median income is {round(correlation_female[0],2)}"
)

(slope, intercept, rvalue, pvalue, stderr) = linregress(
    df_inc_gender["Median Income (dollars)"], df_inc_gender["Total Female"]
)
regress_values_f = [
    income * slope + intercept for income in df_inc_gender["Median Income (dollars)"]
]
line_eq = "y = " + str(slope) + "x + " + str(round(intercept, 2))
# line_eq
print(line_eq)

plt.scatter(
    df_inc_gender["Median Income (dollars)"], df_inc_gender["Total Male"], label="Male"
)
plt.scatter(
    df_inc_gender["Median Income (dollars)"],
    df_inc_gender["Total Female"],
    label="Female",
)
plt.plot(df_inc_gender["Median Income (dollars)"], regress_values_m, "b-")
plt.plot(df_inc_gender["Median Income (dollars)"], regress_values_f, "r-")
plt.title("Median Income vs. Population by Gender")
plt.xlabel("Median Income ($)")
plt.ylabel("Total Population by Gender")
plt.legend()
plt.savefig("med_inc_v_gender.jpg")
plt.show()
```
