# Correlation between Education levels and Health issues

## Background
### The relationship between education and health is never a simple one. Poor health not only results from lower educational attainment, it can also cause educational setbacks and interfere with education. Health conditions like Diabetes have an effect on educational outcomes.

## Observation
### The conclusion drawn from the data set that has been used is that Diabetes is more with higher education within the female population.


```python
# Dependencies
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

%load_ext lab_black
```

```python
# Pull data
df = pd.read_csv("../Data/mrgdata_food.csv")
```

```python
# Copy data
df1 = df.copy()
df1.set_index("Census Tract", inplace=True)
df1.head()
```

```python
# count col & row
df1.shape
```

```python
# Cleaning data
df1 = df1.dropna()
```

```python
# Rechecking the count of col & row
df1.shape
```

### Analysis
#### Relationship between 18+ population and diseases ( Diabetes, Blood Pressure and Obesity) 

### Result
#### The disease that was least was identified which in this case was Diabetes

```python
# create a scatter of Total 18+ Population with different health issues & save as .png
plt.scatter(
    df1["Total 18+ Population"],
    df1["Data_Value_dbt"],
    edgecolors="black",
    facecolors="skyblue",
)
plt.scatter(
    df1["Total 18+ Population"],
    df1["Data_Value_bp"],
    edgecolors="black",
    facecolors="orange",
)
plt.scatter(
    df1["Total 18+ Population"],
    df1["Data_Value_chol"],
    edgecolors="black",
    facecolors="green",
)
plt.scatter(
    df1["Total 18+ Population"],
    df1["Data_Value_obs"],
    edgecolors="black",
    facecolors="red",
)
plt.title("Total no of 18+ Population V different health issues(%)")
plt.xlabel(" Population ")
plt.ylabel(" Diabetes,BP,cholesterol & Obesity(%)")
plt.grid(b=True, which="major", axis="both", linestyle="-", color="lightgrey")

plt.savefig("plot1.png", dpi=100, bbox_inches="tight")
plt.show()
print(
    f"relationship between 18+ population and diseases (Cholesterol, Diabetes, Blood Pressure and Obesity"
)

print(
    f"disease that was least on total 18+population was identified which in this case was Diabetes (BLUE)"
)
```

### Analysis
#### The relationship between diabetes and male/female population 
### Result
#### Diabetes was more in the female population

```python
# create a scatter of Total male & Female 18+ Population with dbt issue
plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total 18+ Male Population"],
    edgecolors="black",
    facecolors="skyblue",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total 18+ Female Population"],
    edgecolors="black",
    facecolors="pink",
)


plt.title("Total 18+male & Female Population, Diabetes")
plt.xlabel("Diabetes(%)")
plt.ylabel("Male & Female Population")

plt.grid(b=True, which="major", axis="both", linestyle="-", color="lightgrey")

plt.savefig("plot2.png", dpi=100, bbox_inches="tight")
plt.show()
print(f"The relationship between diabetes and male/female population ")

print(f"concluded that Diabetes was more in the female population (PINK) ")
```

### Analysis
#### Female with different age group with Diabetes issue 
### Result
#### Diabetes was more in the age group of 65+ female  population

```python
# create a scatter of Female with different age group with dbt issue
plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 18 to 24 years"],
    edgecolors="black",
    facecolors="red",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 25 to 34 years"],
    edgecolors="black",
    facecolors="pink",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Male 35 to 44 years"],
    edgecolors="black",
    facecolors="yellow",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 45 to 64 years"],
    edgecolors="black",
    facecolors="skyblue",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 65 years and over"],
    edgecolors="black",
    facecolors="green",
)


plt.title("Female with various age group , Diabetes")
plt.xlabel("Diabetes(%)")
plt.ylabel("various age group ")
plt.grid(b=True, which="major", axis="both", linestyle="-", color="lightgrey")

plt.savefig("plot3.png", dpi=100, bbox_inches="tight")
plt.show()
print(f"The relationship between diabetes and various age group of female population ")

print(f"concluded that Diabetes was more in the female population with 65+ (GREEN) ")
```

### Analysis
#### The last correlation drawn was between education in female with age group 45 to 64 years population vs diabetes.

### Result
#### The conclusion drawn from the data set that has been used is that Diabetes is more with higher education within the female population.

```python
# create a scatter of Female age group 45 to 64 having different education level with dbt issue
plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 45 to 64 years Less than 9th grade"],
    edgecolors="black",
    facecolors="red",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 45 to 64 years 9th to 12th grade, no diploma"],
    edgecolors="black",
    facecolors="pink",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 45 to 64 years High school graduate"],
    edgecolors="black",
    facecolors="yellow",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 45 to 64 years Some college, no degree"],
    edgecolors="black",
    facecolors="skyblue",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 45 to 64 years Associate's degree"],
    edgecolors="black",
    facecolors="green",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 18 to 24 years Bachelor's degree"],
    edgecolors="black",
    facecolors="purple",
)

plt.scatter(
    df1["Data_Value_dbt"],
    df1["Total Female 45 to 64 years Graduate or professional degree"],
    edgecolors="black",
    facecolors="gray",
)


plt.title(" Feale age group 45 to 64 , Diabetes")
plt.xlabel("Diabetes")
plt.ylabel(" Male age group 18 to 24 having different education level ")
plt.grid(b=True, which="major", axis="both", linestyle="-", color="lightgrey")

plt.savefig("plot4.png", dpi=100, bbox_inches="tight")
plt.show()
print(
    f"The last correlation drawn was between education level in female population vs diabetes"
)

print(
    f"The conclusion drawn from the data set that has been used is that Diabetes is more with higher education within the female population(GRAY) "
)
```

### Analysis
