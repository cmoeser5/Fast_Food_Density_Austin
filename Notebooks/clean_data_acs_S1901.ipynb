```python
import pandas as pd
file_to_load = '../Data/ACSST5Y2017.S1901_data_with_overlays_2020-09-23T184134.csv'
df = pd.read_csv(file_to_load, usecols = ['GEO_ID', 'S1901_C01_001E', 'S1901_C01_002E','S1901_C01_003E', 'S1901_C01_004E', 'S1901_C01_005E', 'S1901_C01_006E', 'S1901_C01_007E', 'S1901_C01_008E', 'S1901_C01_009E', 'S1901_C01_010E', 'S1901_C01_011E', 'S1901_C01_012E', 'S1901_C01_013E'    ])
df.head()
```

```python
df = df.rename(columns={"GEO_ID": "Census_Tract",
                        "S1901_C01_001E": "Total Households",
                        "S1901_C01_002E": "Income < $10,000", 
                       "S1901_C01_003E": "Income $10,000-$14,999",
                       "S1901_C01_004E": "Income $15,000-$24,999",
                       "S1901_C01_005E": "Income $25,000-$34,999",
                       "S1901_C01_006E": "Income $35,000-$49,999",
                       "S1901_C01_007E": "Income $50,000-$74,999",
                       "S1901_C01_008E": "Income $75,000-$99,999",
                       "S1901_C01_009E": "Income $100,000-$149,999",
                       "S1901_C01_010E": "Income $150,000-$199,999",
                       "S1901_C01_011E": "Income > $200,000",
                       "S1901_C01_012E": "Median Income (dollars)",
                       "S1901_C01_013E": "Mean Income (dollars)" 
                       })
df
```

```python
df.drop(df.index[0], inplace=True)
df
```

```python
df.reset_index()
```

```python
df[['GEO_ID', 'Census Tract']] = df.Census_Tract.apply(
    lambda x: pd.Series(str(x).split('US'))
)
df
```

```python
df.drop(columns = ['Census_Tract'], inplace=True)
df
```
