```python
import pandas as pd
```

```python
df_cdc = pd.merge(
    pd.read_csv("Data/cleaned_blood_pressure_cdc.csv"),
    pd.read_csv("Data/cleaned_cholesterol_cdc.csv"),
    on="TractFIPS",
    how="inner",
    suffixes=("_bp", "_chol"),
)

df_cdc
```

```python
df_cdc = pd.merge(
    df_cdc, pd.read_csv("Data/cleaned_diabetes_cdc.csv"), on="TractFIPS", how="inner"
)


df_cdc
```

```python
df_cdc.columns
```

```python
df_cdc = pd.merge(
    df_cdc,
    pd.read_csv("Data/cleaned_obesity_cdc.csv"),
    on="TractFIPS",
    how="inner",
    suffixes=("_dbt", "_obs"),
)


df_cdc
```

```python
df_cdc.columns
```

```python
df_acs = pd.merge(
    pd.read_csv("Data/cleaned_education_acs.csv"),
    pd.read_csv("Data/cleaned_income_acs.csv"),
    on="Census Tract",
    how="inner",
)

df_acs
```

```python
df_acs = pd.merge(
    df_acs, pd.read_csv("Data/cleaned_race_acs.csv"), on="Census Tract", how="inner"
)


df_acs
```

```python
df_cdc = df_cdc.rename(columns={"TractFIPS": "Census Tract"})
df_cdc.columns
```

```python
df_all_data = pd.merge(df_cdc, df_acs, on="Census Tract", how="inner",)

df_all_data
```

```python
df_all_data.columns
```

```python
df_all_data.to_csv("Data/mrgdata.csv")
```
